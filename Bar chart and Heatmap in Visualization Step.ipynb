**Đọc tập tin flight_deplays.csv và lưu vào biến DataFrame tên flights**

```python
flights=pd.read_csv('flight_delays.csv')
flights=flights.set_index('Month')
```

**Chart: Average Arrival Delay for Each Airline, by Month**

```python
plt.figure(figsize=(14,7))
plt.title("Average Arrival Delay for Each Airline, by Month")
sns.heatmap(data=flights, annot=True)
plt.xlabel("Airline")
plt.show()
plt.clf()
```

Biểu đồ: Tổng thời gian đến trể của các chuyến bay trong nằm

```python
flights_=flights.sum().to_frame()
flights_=flights_[flights_[0]>0]
flights_=flights_.sort_values(by=0,ascending=False)
plt.figure(figsize=(8,6))
sns.barplot(data=flights_,x=flights_.index,y=flights_[0],ci=None)
plt.title("Tổng thời gian đến trễ của các chuyến bay trong năm")
plt.ylabel('Thời gian trễ')
plt.xlabel('Hãng Hàng Không')
plt.show()
plt.clf()
```


Nhận xét:
- Hãng Hàng Không NK có tổng thời gian đến trễ cao nhất
- Hãng Hàng Không DL có tổng thời gian đến trễ thấp nhất

Tổng thời gian đến trễ của các chuyến bay trong mỗi tháng

```python
flights_1=flights
flights_1['Total']=flights_1.sum(axis=1)
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=flights_1,x=flights_1.index,y=flights_1['Total'],ci=None)
plt.title("Tổng thời gian đến trễ của các chuyến bay trong mỗi tháng")
plt.ylabel('Tổng thời gian trễ')
plt.xlabel('Tháng')
plt.show()
plt.clf()
```

Nhận xét: 
- Tháng 2 và Tháng 6 có thời gian đến trễ nhiều nhất, Tháng 10 có thời gian đến trễ thấp nhất.
